```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("..\\..\\DataScienceProjects\\titanic\\train.csv")
df.head()
```

```python
df["Cabin"].fillna("N", inplace=True)
df.head()
```

```python
df["Cabin"] = df["Cabin"].apply(lambda x: str(x)[0])
df.head()
```

```python
df.drop(["PassengerId", "Name", "Age", "SibSp", "Parch", "Ticket", "Fare", "Embarked"], axis=1, inplace=True)
df.head()
```

```python
y_train = df["Survived"]
X_train = df.drop("Survived", axis=1)
X_train
```

```python
df_encoded = pd.get_dummies(X_train, columns=['Sex', 'Cabin'], dtype=int)
df_encoded
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(df_encoded, y_train)
```

```python
df_test = pd.read_csv("..\\..\\DataScienceProjects\\titanic\\test.csv")
df_test["Cabin"].fillna("N", inplace=True)
df_test["Cabin"] = df["Cabin"].apply(lambda x: str(x)[0])
df_test.drop(["PassengerId", "Name", "Age", "SibSp", "Parch", "Ticket", "Fare", "Embarked"], axis=1, inplace=True)
df_test_encoded = pd.get_dummies(df_test, columns=['Sex', 'Cabin'], dtype=int)
df_test_encoded
```

```python
y_pred = clf.predict(df_test_encoded)
```

```python
y_pred.size
```

```python
df_test_final = pd.read_csv("..\\..\\DataScienceProjects\\titanic\\test.csv")
df_submission = pd.DataFrame(df_test_final["PassengerId"])
df_submission["Survived"] = y_pred
df_submission.to_csv("Kaggle_Titanic.csv", index=False)
```
